```python
import logging
from smotdm.data.motionx import MotionXDataset
from tqdm.auto import tqdm
import plotly.express as px

logging.basicConfig(level="DEBUG")
```

```python
dataset = MotionXDataset(root_dir="data/MotionX")


frame_counts = [
    {
        'idx': i,
        'frameCount': sample['num_frames'],
    }
    for i, sample in tqdm(enumerate(dataset), total=len(dataset))
]
```

```python
px.line(frame_counts, x="idx", y="frameCount", title='Frame Counts').show()
```
